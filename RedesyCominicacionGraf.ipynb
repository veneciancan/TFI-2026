```python
import networkx as nx
import matplotlib.pyplot as plt

def simular_red_nodos_visual():
    # Tu diccionario original (¡que está perfecto!)
    red = {
        "Nodo 1": ["Nodo 2", "Nodo 3"],
        "Nodo 2": ["Nodo 3", "Nodo 4"],
        "Nodo 3": ["Nodo 4", "Nodo 5"],
        "Nodo 4": ["Nodo 5", "Nodo 6"],
        "Nodo 5": ["Nodo 6", "Nodo 7"],
        "Nodo 6": ["Nodo 7", "Nodo 8"],
        "Nodo 7": ["Nodo 8", "Nodo 1"],
        "Nodo 8": ["Nodo 1", "Nodo 2"]
    }

    print("--- Generando Simulación Visual de Red de 8 Nodos ---")

    # 1. Crear un grafo DIRIGIDO (DiGraph) por la "relación direccionada"
    G = nx.DiGraph()

    # 2. Recorrer tu diccionario para agregar las conexiones (aristas/edges) al grafo
    for origen, destinos in red.items():
        for destino in destinos:
            G.add_edge(origen, destino)

    # 3. Dibujar el grafo
    # Usamos circular_layout porque para redes de este estilo queda muy ordenado visualmente
    pos = nx.circular_layout(G)

    # Dibujamos con flechas (arrows=True) para mostrar la dirección
    nx.draw(G, pos, with_labels=True, node_color='lightgreen',
            node_size=2000, font_size=10, font_weight='bold', arrows=True)

    plt.show()

if __name__ == "__main__":
    simular_red_nodos_visual()
```
